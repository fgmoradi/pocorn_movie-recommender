```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score

import pickle
```

```python
nltk.download("stopwords")
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/datasets/reviews.txt',sep = '\t', names =['Reviews','Comments'])
```

```python
dataset
```

```python
stopset = set(stopwords.words('english'))
```

```python
vectorizer = TfidfVectorizer(use_idf = True,lowercase = True, strip_accents='ascii',stop_words='english')
```

```python
X = vectorizer.fit_transform(dataset.Comments)
y = dataset.Reviews
pickle.dump(vectorizer, open('Mytranform.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X,y)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
y_pred = clf.predict(X_test)

precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
filename = 'My_model.pkl'
pickle.dump(clf, open(filename, 'wb'))
```
